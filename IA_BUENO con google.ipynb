```python
# Convertir columnas a min칰sculas para evitar errores
df.columns = [col.strip().lower() for col in df.columns]
```

```python
# Buscar modelos por serie y mostrar tabla con serie + modelo
def buscar_modelos_por_serie(serie):
    serie = serie.strip().lower()

    # Filtrar registros que contengan el nombre parcial de la serie
    resultados = df[df['serie'].str.lower().str.contains(serie, na=False)]

    if resultados.empty:
        return f"No se encontraron modelos que pertenezcan a una serie que contenga: '{serie}'."

    # Seleccionar solo columnas deseadas y eliminar duplicados
    tabla = resultados[['serie', 'modelo']].drop_duplicates()

    # Devolver el DataFrame directamente para que Colab lo muestre en formato tabla
    return tabla.reset_index(drop=True)
```

```python
# Buscar modelos por serie y mostrar tabla con serie + modelo
def buscar_numero_material(modelo):
    modelo = modelo.strip().lower()

    # Filtrar registros que contengan el nombre parcial de la serie
    resultados = df[df['modelo'].str.lower().str.contains(modelo, na=False)]

    if resultados.empty:
        return f"No se encontraron numeros de materiales que pertenezcan al modelo: '{modelo}'."

    # Seleccionar todas las columnas deseadas en una sola tabla
    columnas = ['modelo', 'serie', 'numero material']
    columnas_existentes = [col for col in columnas if col in df.columns]

    # Un solo DataFrame
    return resultados[columnas_existentes].drop_duplicates().reset_index(drop=True)
```

```python
# Caracter칤sticas de uno o varios modelos con formato de moneda en el precio
def caracteristicas_modelo(nombre_modelo):
    nombre_modelo = nombre_modelo.strip().lower()

    # Filtrar modelos que contengan ese texto
    coincidencias = df[df['modelo'].str.lower().str.contains(nombre_modelo, na=False)]

    if coincidencias.empty:
        return f"No se encontraron modelos que contengan '{nombre_modelo}'."

    columnas_deseadas = ['modelo', 'serie', 'aumento', 'dimensiones', 'peso (g)', 'n췈 material', 'precio']
    columnas_existentes = [col for col in columnas_deseadas if col in df.columns]
    tabla = coincidencias[columnas_existentes].copy()

    # 游녤 Formatear la columna "precio" si existe
    if 'precio' in tabla.columns:
        tabla['precio'] = tabla['precio'].apply(lambda x: f"{x} " if pd.notnull(x) else x)

    return tabla.reset_index(drop=True)
```

```python
import re

def precio_de_modelo(filtro):
    filtro = filtro.strip().lower()

    # Si se solicita ver todos los productos
    if filtro == "" or filtro == "todos":
        resultados = df[['modelo', 'precio']].dropna().copy()

    # Si es un rango tipo "entre 100 y 200"
    elif ("entre" in filtro and "y" in filtro) or ("entre" in filtro and "-" in filtro):
        numeros = re.findall(r"\d+(?:[.,]\d+)?", filtro)
        if len(numeros) == 2:
            min_val = float(numeros[0].replace(",", "."))
            max_val = float(numeros[1].replace(",", "."))
            resultados = df[(df['precio'] >= min_val) & (df['precio'] <= max_val)][['modelo', 'precio']].copy()
        else:
            return "No entend칤 bien el rango. Usa por ejemplo: 'entre 100 y 300'"

    # Si es tipo "mayor de 300"
    elif "mayor de" in filtro:
        numero = re.search(r"\d+(?:[.,]\d+)?", filtro)
        if numero:
            val = float(numero.group().replace(",", "."))
            resultados = df[df['precio'] > val][['modelo', 'precio']].copy()
        else:
            return "Indica un n칰mero v치lido despu칠s de 'mayor de'."

    # Si es tipo "menor de 300"
    elif "menor de" in filtro:
        numero = re.search(r"\d+(?:[.,]\d+)?", filtro)
        if numero:
            val = float(numero.group().replace(",", "."))
            resultados = df[df['precio'] < val][['modelo', 'precio']].copy()
        else:
            return "Indica un n칰mero v치lido despu칠s de 'menor de'."

    # Si es b칰squeda por nombre parcial del modelo
    else:
        resultados = df[df['modelo'].str.lower().str.contains(filtro, na=False)][['modelo', 'precio']].copy()

    # Si no se encuentra nada
    if resultados.empty:
        print ("Prueba con:")
        print("- entre 100 y 200")
        print("- mayor de 300")
        print("- menor de 250")
        print("- DTC")
        print("- Todos")

    # Ordenar y dar formato
    resultados = resultados.drop_duplicates().sort_values(by='precio')
    resultados['precio'] = resultados['precio'].apply(lambda x: f"{x:,.2f} " if pd.notnull(x) else x)

    return resultados.reset_index(drop=True)
```

```python
# Buscar modelos por serie y mostrar tabla con serie + modelo
def dimensiones_de_modelo(nombre_modelo):
    nombre_modelo = nombre_modelo.strip().lower()

    # Filtrar registros que contengan el nombre parcial de la serie
    resultados = df[df['modelo'].str.lower().str.contains(nombre_modelo, na=False)]

    if resultados.empty:
        return f"No se encontraron modelos que pertenezcan a una serie que contenga: '{nombre_modelo}'."

    # Seleccionar solo columnas deseadas y eliminar duplicados
    tabla = resultados[['dimensiones', 'modelo', 'serie']].drop_duplicates()

    # Devolver el DataFrame directamente para que Colab lo muestre en formato tabla
    return tabla.reset_index(drop=True)
```

```python
def responder_peso(pregunta):
    pregunta = pregunta.strip().lower()

    if 'modelo' not in df.columns or 'peso (g)' not in df.columns:
        return "Las columnas necesarias no est치n disponibles en el DataFrame."

    # Caso: entre X y Y
    if "entre" in pregunta and "y" in pregunta:
        numeros = re.findall(r"\d+(?:[.,]\d+)?", pregunta)
        if len(numeros) == 2:
            min_val = float(numeros[0].replace(",", "."))
            max_val = float(numeros[1].replace(",", "."))
            resultados = df[(df['peso (g)'] >= min_val) & (df['peso (g)'] <= max_val)][['modelo', 'peso (g)']].copy()
        else:
            return "No se entendi칩 bien el rango. Usa por ejemplo: 'entre 300 y 600'"

    # Caso: mayor de X
    elif "mayor de" in pregunta:
        num = re.search(r"\d+(?:[.,]\d+)?", pregunta)
        if num:
            val = float(num.group().replace(",", "."))
            resultados = df[df['peso (g)'] > val][['modelo', 'peso (g)']].copy()
        else:
            return "Indica un n칰mero v치lido despu칠s de 'mayor de'."

    # Caso: menor de X
    elif "menor de" in pregunta:
        num = re.search(r"\d+(?:[.,]\d+)?", pregunta)
        if num:
            val = float(num.group().replace(",", "."))
            resultados = df[df['peso (g)'] < val][['modelo', 'peso (g)']].copy()
        else:
            return "Indica un n칰mero v치lido despu칠s de 'menor de'."

    # Caso general: modelo por texto parcial
    else:
        resultados = df[df['modelo'].str.lower().str.contains(pregunta, na=False)][['modelo', 'peso (g)']].copy()

    if resultados.empty:
        return f"No se encontraron modelos relacionados con '{pregunta}'."

    # Formatear peso con "g"
    resultados['peso (g)'] = resultados['peso (g)'].apply(lambda x: f"{x} g" if pd.notnull(x) else x)
    resultados = resultados.drop_duplicates().sort_values(by='peso (g)')

    return resultados.reset_index(drop=True)
```

```python
def buscar_pesos(modelo):
    modelo = modelo.strip().lower()

    # Verificar que las columnas existen
    if 'serie' not in df.columns or 'modelo' not in df.columns or 'peso (g)' not in df.columns:
        return "Las columnas necesarias no est치n disponibles en el DataFrame."

    # Filtrar por coincidencia parcial del modelo
    resultados = df[df['modelo'].str.lower().str.contains(modelo, na=False)]

    if resultados.empty:
        return f"No se encontraron modelos que contengan '{modelo}'."

    # Seleccionar y formatear columnas
    tabla = resultados[['serie', 'modelo', 'peso (g)']].dropna().drop_duplicates()
    tabla['peso (g)'] = tabla['peso (g)'].apply(lambda x: f"{x} g" if pd.notnull(x) else x)

    return tabla.reset_index(drop=True)
```

```python
# Introducimos el chatbot y toda la informaci칩n que queremos que nos muestre cuando se ejecute esta funci칩n
def chatbot():
    print("游뱄 춰Hola! Soy tu asistente de Minox.")
    print("Puedes preguntarme cosas como:")
    print("- Modelo de la serie (ej:'modelo de la serie RS-4')")
    print("- Caracteristicas del modelo (ej: 'Modelo DTC 1200')")
    print("- Precio de los productos (ej: 'Precio de DTC 1200')")
    print("- Dimensiones de los produstos (ej: 'Dimensiones de DTC 1200)")
    print("- Peso de los productos (ej: 'Peso de DTC 1200)")
    print("- N칰mero de material de los productos (ej: 'N칰mero de material de DTC 1200)")
    print("Escribe 'salir' para terminar.\n")
```

```python
# Funci칩n principal para iniciar el chatbot
def iniciar_chatbot():
    # Llama a la funci칩n chatbot()
    chatbot()

    # Bucle infinito para mantener la conversaci칩n hasta que el usuario escriba "salir"
    while True:
        try:
            # Solicita al usuario una pregunta, la convierte a min칰sculas para facilitar la comparaci칩n
            pregunta = input("T칰: ").lower()

            # Si el usuario quiere salir, se rompe el bucle
            if pregunta == "salir":
                print("游녦 춰Hasta luego!")
                break

            # Si la pregunta contiene palabras clave relacionadas con los modelos de una serie
            elif any(p in pregunta for p in ["modelo de la serie", "modelos de la serie", "modelos", "modelo"]):
                serie = input("游댳 쯅ombre de la serie?: ")
                from IPython.display import display  # Importaci칩n local para mostrar resultados en notebooks
                display(buscar_modelos_por_serie(serie))  # Muestra los modelos de la serie especificada

            # Si la pregunta es sobre precios
            elif any(p in pregunta for p in ["precio", "cu치nto", "vale", "coste", "costar칤a"]):
                nombre = input("游댳 쯅ombre del modelo?: ")
                from IPython.display import display
                display(precio_de_modelo(nombre))  # Muestra el precio del modelo

            # Si la pregunta es sobre dimensiones
            elif any(p in pregunta for p in ["dimensiones", "dimensiones de", "medidas", "medida", "dimensi칩n"]):
                modelo = input("游댳 쯅ombre del modelo?: ")
                from IPython.display import display
                display(dimensiones_de_modelo(modelo))  # Muestra las dimensiones del modelo

            # Si la pregunta es sobre caracter칤sticas generales o materiales
            elif any(p in pregunta for p in ["caracteristicas", "materiales", "informacion", "caracteristica"]):
                modelo = input("游댳 쯅ombre del modelo?: ")
                from IPython.display import display
                display(caracteristicas_modelo(modelo))  # Muestra informaci칩n del modelo

            # Si la pregunta es sobre el n칰mero de material (c칩digo SAP)
            elif any(p in pregunta for p in ["numero material", "n칰mero de material", "sap", "n칰meros materiales"]):
                modelo = input("游댳 쯅ombre del modelo?: ")
                from IPython.display import display
                display(buscar_numero_material(modelo))  # Muestra el n칰mero de material

            # Si la pregunta es sobre el peso
            elif any(p in pregunta for p in ["peso material", "peso", "pesos", "peso modelos"]):
                modelo = input("游댳 쯅ombre del modelo?: ")
                from IPython.display import display
                display(responder_peso(modelo))  # Muestra el peso del modelo

            # Si no se reconoce la intenci칩n de la pregunta
            else:
                print("Lo siento, no entend칤 tu pregunta. Intenta reformularla.")

        # Captura cualquier error durante la ejecuci칩n para evitar que se interrumpa el bucle
        except Exception as e:
            print(f"丘멆잺 Se produjo un error: {e}")
            continue
```

```python
# Ejecutamo el chatbot y le hacemos las preguntas
iniciar_chatbot()
```

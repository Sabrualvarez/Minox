```python
# Paso 4: Limpieza de datos
# Elimina columnas completamente vacías
dataframe = dataframe.dropna(axis=1, how='all')

# Elimina filas completamente vacías
dataframe = dataframe.dropna(axis=0, how='all')

# Elimina espacios en nombres de columnas
dataframe.columns = dataframe.columns.str.strip()

# Opcional: Elimina duplicados
dataframe = dataframe.drop_duplicates()

# Opcional: Limpia la columna 'Precio' si hay símbolo € o comas
if 'Precio' in dataframe.columns:
    # Convertir a string y limpiar símbolos
    dataframe['Precio'] = dataframe['Precio'].astype(str).apply(
        lambda x: str(x).replace('€', '').replace('.', '').replace(',', '.').strip()
        if isinstance(x, str) else x
    )
    # Convertir a número (float)
    dataframe['Precio'] = pd.to_numeric(dataframe['Precio'], errors='coerce')


# Paso 5: Guarda el archivo limpio
dataframe.to_excel("/content/datos_limpiados.xlsx", index=False)

# Paso 6: Descarga el archivo (opcional)
from google.colab import files
files.download("/content/datos_limpiados.xlsx")
```
